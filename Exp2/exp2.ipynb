### 1. Importing Libraries and Dataset

### 3. Further Steps

```python
# load
df = pd.read_csv('../Datasets/maharashtra-census-2011.csv')
df.head()
```

```python
# print all columns
print(df.columns)
```

```python
# Word Chart - Most frequent District names
district_counts = df['District name'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=district_counts.index, y=district_counts.values, hue=district_counts.index, legend=False)
plt.title("Word Chart: Frequency of District Names")
plt.xlabel("District Name")
plt.ylabel("Frequency")
plt.xticks(rotation=90)
plt.show()
print("Observation: All district names in Maharashtra are unique, so the frequency is 1 for each.")
```

```python
# Violin Plot - Distribution of Male and Female Population
plt.figure(figsize=(10, 6))
sns.violinplot(data=df[['Male', 'Female']], palette="Set2")
plt.title("Violin Plot: Male and Female Population Distribution")
plt.show()
print("Observation: The distributions of male and female populations are similar, but males have a slightly wider spread.")
```

```python
# Regression Plot - Literate vs. Population (Linear)
plt.figure(figsize=(10, 6))
sns.regplot(x='Population', y='Literate', data=df)
plt.title("Regression Plot: Population vs. Literate (Linear)")
plt.show()
print("Observation: There is a strong positive linear relationship between the population and the literate population.")
```

```python
# Regression Plot - Literate vs. Population (Nonlinear)
plt.figure(figsize=(10, 6))
sns.regplot(x='Population', y='Literate', data=df, order=2)
plt.title("Regression Plot: Population vs. Literate (Nonlinear)")
plt.show()
print("Observation: The nonlinear regression also shows a positive relationship, with some slight curvature indicating diminishing returns.")
```

```python
# 3D Chart - Population, Literate, and Workers
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Population'], df['Literate'], df['Workers'], c='b', marker='o')
ax.set_xlabel('Population')
ax.set_ylabel('Literate')
ax.set_zlabel('Workers')
plt.title("3D Chart: Population vs. Literate vs. Workers")
plt.show()
print("Observation: There is a positive correlation between population, literacy, and the number of workers.")
```

```python
# Jitter Plot - Literacy vs. Education Levels
plt.figure(figsize=(10, 6))
sns.stripplot(x='Literate', y='Higher_Education', data=df, jitter=True, hue='Higher_Education', legend=False)
plt.title("Jitter Plot: Literate vs. Higher Education Levels")
plt.show()
print("Observation: Districts with higher literacy tend to have higher levels of higher education.")
```

```python
# Area Plot - Education Levels by District
df.set_index('District name', inplace=True)
df[['Below_Primary_Education', 'Primary_Education', 'Middle_Education', 'Secondary_Education', 
    'Higher_Education']].plot(kind='area', stacked=True, figsize=(12, 6), alpha=0.5)
plt.title("Area Plot: Education Levels by District")
plt.ylabel('Number of People')
plt.show()
print("Observation: Primary education is the most prevalent, followed by middle and secondary education levels.")
```

```python
# Waterfall Plot - Change in Population by District
districts = df['Population'].sort_values(ascending=False).index
population_diff = df['Population'].sort_values(ascending=False).diff().fillna(df['Population'].max())

waterfall(districts, population_diff, threshold=0.05)
plt.title("Waterfall Plot: Change in Population by District")
plt.show()
print("Observation: The plot shows the difference in population between consecutive districts.")
```

```python
# Donut Chart - Religion Distribution in a District (Example: First District)
religion_cols = ['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains', 'Others_Religions', 'Religion_Not_Stated']
religion_data = df.loc[df.index[0], religion_cols]

plt.figure(figsize=(8, 8))
plt.pie(religion_data, labels=religion_cols, autopct='%1.1f%%', startangle=140)
plt.gca().add_artist(plt.Circle((0,0), 0.70, color='white'))
plt.title("Donut Chart: Religion Distribution in District - " + df.index[0])
plt.show()
print("Observation: The majority religion in the district is Hindus, with other religions contributing smaller percentages.")
```

```python
sizes = df['Population'].values
labels = df['District name']
plt.figure(figsize=(12, 8))
squarify.plot(sizes=sizes, label=labels, alpha=.8)
plt.title("Treemap: Population Distribution by District")
plt.axis('off')
plt.show()
print("Observation: The treemap highlights the population distribution across districts, showing a few districts dominate in terms of population.")
```

```python
# Funnel Plot - Literacy Funnel
literacy_funnel = df[['Literate', 'Male_Literate', 'Female_Literate']].sum()

plt.figure(figsize=(8, 6))
sns.barplot(x=literacy_funnel.values, y=literacy_funnel.index, hue=literacy_funnel.index, legend=False)
plt.title("Funnel Plot: Literacy Levels")
plt.show()
print("Observation: The majority of the literate population is male, with a smaller but significant female literate population.")
```
