### 1. Importing Libraries

```python
import pandas as pd
import numpy as np
```

### 2. Loading Dataset

```python
df = pd.read_csv('../Datasets/crime.csv', encoding='ISO-8859-1')
df.head()
```

### 3. Data Preprocessing

```python
df.isnull().sum()
```

```python
# drop columns: offense_code_extension, last_occurrence_date, geo_lon, geo_lat, precinct_id
df.drop(['offense_code_extension', 'last_occurrence_date', 'geo_lon', 'geo_lat', 'precinct_id'], axis=1, inplace=True)

# drop rows with missing values
df.dropna(inplace=True)
```

```python
print(df.shape)
df.head()
```

```python
# Save the file with UTF-8 encoding
df.to_csv("../Datasets/crime_cleaned.csv", encoding="utf-8", index=False)
```

- **Scatter Plot** - To show the no. of crimes by neighborhood
